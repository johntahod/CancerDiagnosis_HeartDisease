### A : Read dataset



```python
import numpy as np
import pandas as pd


df = pd.read_csv('https://github.com/mpourhoma/CS4661/raw/master/Cancer.csv')
# df[0::10]
```

### B : b- Use sklearn functions to split the dataset into testing and training sets with the following parameters: test_size=0.35, random_state=3.

```python
feature_cols = ['Clump_Thickness', 'Uniformity_of_Cell_Size', 'Uniformity_of_Cell_Shape', 'Marginal_Adhesion',
                'Single_Epithelial_Cell_Size', 'Bare_Nuclei', 'Bland_Chromatin', 'Normal_Nucleoli', 'Mitoses']


                


# use the above list to select the features from the original DataFrame
X = df[feature_cols]  

y = df['Malignant_Cancer']

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.35, 
                                                    random_state=3)
```

###  C  : Using Decision Tree 

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score


my_decisiontree = DecisionTreeClassifier(random_state=3)

my_decisiontree.fit(X_train, y_train)

y_predict_dt = my_decisiontree.predict(X_test)


accuracy_dt = accuracy_score(y_test, y_predict_dt)
print('Accuracy of Decision Tree : ',accuracy_dt)
```

### D : Bagging on Voting with 19 Decision Tree Classifiers

```python
# Import

# # Create prediction scores list
# For loop for 19 decision tree classifier portion
#  A “for” loop with loop variable i = 0...18.
#make a bootstarp sample of the original “Training” Dataset (build in part(b)) with the 
# size of bootstarp_size = 0.8*(Size of the original dataset). 
# Testing on the testing set using Base_DescisionTree on X_test from Part B(Part 3)


## This takes in the first place increasing until the end



from sklearn.utils import resample

# Create prediction scores list
predictions = []



for i in range(19):
    
    # Set bootstrap size
    bootstrap_size = int(0.8*df.shape[0])
    
    X_bag, y_bag = resample(X_train, y_train, n_samples=bootstrap_size, random_state=i, replace=True)
    
    Base_DecisionTree  = DecisionTreeClassifier(random_state=3)
    
   # Training ONLY on the training set:
    Base_DecisionTree.fit(X_bag, y_bag)
    
    
    y_predict = Base_DecisionTree.predict(X_test)

    predictions.append(y_predict)

# voting part   
voting = []
for i in range(53): # Range of each sublist
    zeros = 0
    ones = 0;
    for j in range(19):
        number =predictions[j][i] 
        if(number == 0):
            zeros +=1
        else:
            ones +=1
    if(zeros>ones):
        voting.append(0)
    else:
        voting.append(1)

# Check voting list AND calculte accuracy on final results and display
print(voting)
accuracy = accuracy_score(y_test, voting)
print('Accuracy:', accuracy)
```

### E : Random Forest Classifier for Prediction with part B 

```python
from sklearn.ensemble import RandomForestClassifier

my_RandomForest = RandomForestClassifier(n_estimators = 19, bootstrap = True, random_state=3)
my_RandomForest.fit(X_train,y_train)

y_predict_rf = my_RandomForest.predict(X_test)


from sklearn.metrics import accuracy_score

score_rf = accuracy_score(y_test, y_predict_rf)

print(score_rf)
```
